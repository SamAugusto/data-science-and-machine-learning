```python
import numpy as np
import pandas as pd
```

```python
numbers_df = pd.DataFrame(np.arange(0,90,3).reshape(10,3),index = ['row 1','row 2','row 3','row 4','row 5','row 6','row 7','row 8','row 9','row 10'],
                       columns = ['column 1','column 2','column 3'])
numbers_df
```

```python
numbers_df.iloc[0,1]
```

```python
numbers_df.iloc[0,1] = 20
numbers_df.iloc[0,1]
```

```python
numbers_df.iloc[[1,2,4],[1,2]]
```

```python
mask = numbers_df >30
mask
```

```python
numbers_df[mask]
```
