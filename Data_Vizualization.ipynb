```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
```

```python

data = {'names': ['steve','john','richard','sarah','randy','micheal' ,'julie'],
        'age': [20,22,20,21,24,23,22],
        'gender':['Male','Male','Male','Female','Make','Male','Female'],
        'rank':[2,1,4,5,3,7,6]}
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

```python
plot = sns.barplot(data= df, x= 'names',y='age')
plot.set_title("Comparing Ages")
plt.show()
```

# Line plot Matplotlib

```python
plt.plot(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

# Line Plot Seaborn

```python
plot = sns.lineplot(data =df, x= 'names', y = 'age')
plot.set_title("Comparing Ages")
plt.show()
```

#Pie Chart Matplot lib

```python
plt.pie(df['age'],labels = df['names'])
plt.title('Age Comparison')
plt.show()
```

#Pie Char Seaborn

```python
colors = sns.color_palette('pastel')[0:5]
plt.pie(df['age'],labels = df['names'],colors = colors)
plt.show()
```

```python
x =  range(1,10)
y = [1,2,3,4,0,4,3,2,1]
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.set_xlim([1,9])
ax.set_ylim([0,5])
ax.set_xticks(range(11))
ax.set_yticks(range(6))
ax.plot(x,y)
```

```python
x =  range(1,10)
y = [1,2,3,4,0,4,3,2,1]
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.set_xlim([1,9])
ax.set_ylim([0,5])
ax.grid()
ax.plot(x,y)
```

# Creating Statistical graphics in Seaborn

## Eyeballing dataset distributions with histograms

```python
address = 'mtcars.csv'
cars = pd.read_csv(address)
cars.columns= ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs',
               'am','gear','carb']
cars.index = cars.car_names
mpg = cars['mpg']
mpg.plot(kind = 'hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sb.displot(mpg)
```

## Seeing  scatterplots in action

```python
cars.plot(kind = 'scatter',x = 'hp',y= 'mpg',c=['darkgray'],s=150)
```

```python
sb.regplot(x = 'hp',y= 'mpg',data = cars,scatter = True)
```

## Generating a scatter plot matrix

```python
sb.pairplot(cars)
```

```python
cars_subset= cars[['mpg','hp','wt']]
sb.pairplot(cars_subset)
```

## Building boxplots
