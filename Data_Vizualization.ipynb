```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
```

```python

data = {'names': ['steve','john','richard','sarah','randy','micheal' ,'julie'],
        'age': [20,22,20,21,24,23,22],
        'gender':['Male','Male','Male','Female','Make','Male','Female'],
        'rank':[2,1,4,5,3,7,6]}
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

```python
plot = sns.barplot(data= df, x= 'names',y='age')
plot.set_title("Comparing Ages")
plt.show()
```

# Line plot Matplotlib

```python
plt.plot(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

# Line Plot Seaborn

```python
plot = sns.lineplot(data =df, x= 'names', y = 'age')
plot.set_title("Comparing Ages")
plt.show()
```

#Pie Chart Matplot lib

```python
plt.pie(df['age'],labels = df['names'])
plt.title('Age Comparison')
plt.show()
```

#Pie Char Seaborn

```python
colors = sns.color_palette('pastel')[0:5]
plt.pie(df['age'],labels = df['names'],colors = colors)
plt.show()
```

## Creating a line chart from a list object

```python
import numpy as np
from numpy.random import randn
```

```python
address = "mtcars.csv"
cars = pd.read_csv(address)
cars.columns= ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs',
               'am','gear','carb']
mpg = cars['mpg']
```

```python
x = range(1,10)
y =[1,2,3,4,0,4,3,2,1]
plt.plot(x,y)
plt.show()
```

### Plotting a line chart from a pandas object

```python
mpg.plot()
```

```python
df = cars[['cyl','wt','mpg']]
df.plot()
```
