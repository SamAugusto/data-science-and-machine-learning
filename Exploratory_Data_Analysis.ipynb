# Pearson correlation analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams
import scipy
from scipy.stats import pearsonr, spearmanr
```

```python
from ast import increment_lineno
%matplotlib inline
rcParams['figure.figsize']= 8,4
sb.set_style('whitegrid')
```

```python
sb.pairplot(cars)
```

```python
x = cars[['mpg','hp','qsec','wt']]
sb.pairplot(x)
```

## Using scipy to calculate the pearson correlation coefficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt =cars['wt']
pearsonr_coefficient,p_value = pearsonr(mpg,hp)
print(f'PearsonR Corrrelation coefficient {pearsonr_coefficient:.3f}')
pearsonr_coefficient,p_value = pearsonr(mpg,qsec)
print(f'PearsonR Corrrelation coefficient {pearsonr_coefficient:.3f}')
pearsonr_coefficient,p_value = pearsonr(mpg,wt)
print(f'PearsonR Corrrelation coefficient {pearsonr_coefficient:.3f}')
```

### Using pandas to calculater pearson r coeeffictien

```python
corr = x.corr()
corr
```

## Using seaborn to visualize the pearson correlation coefficient

```python
sb.heatmap(corr, xticklabels= corr.columns.values,yticklabels=corr.columns.values)
```

# Spearman rank correlation and Chi-square

```python
x = cars[['cyl','vs','am','gear']]
sb.pairplot(x)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']


spearmanr_coefficient,p_value = spearmanr(cyl,vs)
print(f"Spearmanrank correlation Coefficient {spearmanr_coefficient:.3f}")
spearmanr_coefficient,p_value = spearmanr(cyl,am)
print(f"Spearmanrank correlation Coefficient {spearmanr_coefficient:.3f}")
spearmanr_coefficient,p_value = spearmanr(cyl,gear)
print(f"Spearmanrank correlation Coefficient {spearmanr_coefficient:.3f}")
```

## Chi-square test for independence

```python
table = pd.crosstab(cyl,am)
from scipy.stats import chi2_contingency
```

```python
table = pd.crosstab(cyl,am)
chi2, p ,dof, expected = chi2_contingency(table.values)
print(f"Chi-square statistic {chi2:.3f} and p_value {p:.3f}")
```

```python
table = pd.crosstab(cyl,gear)
chi2, p ,dof, expected = chi2_contingency(table.values)
print(f"Chi-square statistic {chi2:.3f} and p_value {p:.3f}")
```

```python
table = pd.crosstab(cyl,vs)
chi2, p ,dof, expected = chi2_contingency(table.values)
print(f"Chi-square statistic {chi2:.3f} and p_value {p:.3f}")
```

## Identifying outliears from tukey boxplots

```python
# Convert all 4 feature columns to float
for col in df.columns[:4]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.iloc[:, 0:4].boxplot(return_type='dict')
plt.title('Boxplot of Iris Dataset Features')
plt.ylabel('cm')
plt.show()
```
