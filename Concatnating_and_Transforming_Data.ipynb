# Concatenating Data

```python
import numpy as np
import pandas as pd
```

```python
DF_obj =  pd.DataFrame(np.arange(36).reshape(6,6))
DF_obj
```

```python
DF_obj_2 = pd.DataFrame(np.arange(15).reshape(5,3))
DF_obj_2
```

```python
pd.concat([DF_obj,DF_obj_2], axis = 1)
```

```python
pd.concat([DF_obj,DF_obj_2], axis = 0)
```
